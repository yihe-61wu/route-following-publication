```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.transforms as mtransforms

plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 7.5
plt.rcParams['axes.titlesize'] = 7.5

figw2 = 3.307
figw1 = 6.85


def complete_figure(fig, axes, index):    
    trans = mtransforms.ScaledTranslation(0.02, -0.02, fig.dpi_scale_trans)
    for label, ax in zip(index, axes.flatten()):
        ax.text(0.0, 1.0, '({})'.format(label), transform=ax.transAxes + trans,
                verticalalignment='top', fontfamily='Arial')
```

```python
def capacity(p, N, k):
    m = (np.log(1 - p**(1/N)) - np.log(k)) / (np.log(1 - k))
    return m
```

```python
p = np.logspace(-4, 0, num=100, endpoint=False)
k = 0.1
```

```python
fig, ax = plt.subplots(figsize=(figw2, 2), dpi=300)

ax.plot(p, capacity(p, 32000, k), c='k', label='$N_{KC}=32000$ in simulation')
ax.plot(p, capacity(p, 20000, k), c='k', ls='dashed', label='$N_{KC}=20000$ in real world')
ax.set_xscale('log')
ax.grid()
ax.legend()
ax.set_ylabel('memory capacity $m$')
ax.set_xlabel('confusion rate $p_{error}$')
# ax.xaxis.set_tick_params(labelsize=6)
# ax.yaxis.set_tick_params(labelsize=6)
```
